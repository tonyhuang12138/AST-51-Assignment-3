Here we can see some data for assignment 3. There are two data files here. First, we'll play around with the lightcurve. This data file has three columns: time, flux and flux error.

This light curve shows the flux of the eclipsing binary CM Draconis over time. The light curve has various features to it:

1) Primary and secondary eclipses of the two stars over a 1.268 day period
2) Rougly sinusoidal modulations of the light curve due to star spots on a rotating stellar surface
3) Stellar flares, which are a sharp rise then slow decay of flux
4) Instrumental effects that affect the flux over time

Let's first plot the light curve to see how it looks

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lightcurve = np.genfromtxt('lightcurve.csv',delimiter=',')
lc_time = lightcurve[:,0]
lc_flux = lightcurve[:,1]
lc_err = lightcurve[:,2]
```

```python
fig = plt.figure()
plt.scatter(lc_time, lc_flux,s=1,color='blue') # you could do errorbars to include the flux error but I don't for now
plt.xlabel('Time (BJD - 2,457,000)')
plt.ylabel('Normalised Flux')
plt.show(block=False)
```

Let's zoom in on a few things to see some different events. First, let's see some primary and secondary eclipses. Note the separation between them and the slighty different heights!

```python
fig = plt.figure()
plt.scatter(lc_time, lc_flux,s=1,color='blue')
plt.xlabel('Time (BJD - 2,457,000)')
plt.ylabel('Normalised Flux')
plt.xlim([1945,1950])
plt.show(block=False)
```
